```python
import matplotlib.pyplot as plt
import numpy as np

# 
diseases = ['Diabetes', 'Hypertension', 'Cancer', 'Heart Disease', 'Alzheimer\'s Disease',
            'Autoimmune Diseases', 'Liver Disease', 'Infectious Diseases', 'Respiratory Diseases', 'Kidney Disease']

# 
effectiveness_nano_robot_3 = [95, 82, 65, 77, 72, 89, 80, 75, 92, 87]  # 
effectiveness_nano_robot_2 = [88, 80, 62, 75, 70, 80, 79, 66, 85, 75]
effectiveness_nano_robot_1 = [82, 66, 42, 62, 55, 72, 66, 63, 77, 68]
effectiveness_traditional =  [72, 52, 33, 55, 49, 62, 58, 52, 65, 53]  # 

# X
x = np.arange(4)  # 0 for Nano Robot, 1 for Traditional Treatment

# 
colors = plt.cm.tab10.colors  # tab10

# 
plt.figure(figsize=(10, 6))

# 
for i, disease in enumerate(diseases):
    plt.plot(x, [effectiveness_nano_robot_3[i], effectiveness_nano_robot_2[i], effectiveness_nano_robot_1[i], effectiveness_traditional[i]], marker='o', color=colors[i % len(colors)], label=disease)

# X
plt.xticks(x, ['BioCapsule-3', 'BioCapsule-2', 'BioCapsule-1', 'Traditional Treatment'])

# Y
plt.ylabel('Cure Rate (%)')

# 
plt.title('Cure Rate Comparison: BioCapsule vs Traditional Treatment')

# 
plt.legend(title='Diseases', bbox_to_anchor=(1.05, 1), loc='upper left')

# 
plt.tight_layout()
plt.grid(True)
plt.show()
```
